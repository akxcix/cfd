```python
%matplotlib widget

from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt
import math

x = np.linspace(0, 1, 300)

CS = 0.2969
C1 = -0.1260
C2 = -0.3516
C3 = 0.2843
C4 = -0.1015

def f(x, cs, c1, c2, c3, c4):
    return cs * np.sqrt(x) + c1 * x + c2 * x ** 2 + c3 * x ** 3 + c4 * x ** 4

def f_(x):
    return f(x, CS, C1, C2, C3, C4)

def g(x, a, b, c, d, e):
    return x * (x - 1) * (a * x ** 0.5 + b + c * x + d * x ** 2 + e * x ** 3)

def g_(x, a, b, c, d, e, add):
    gv = g(x, a, b, c, d, e)
    fv = f_(x)

    if add:
        return gv + fv
    else:
        return gv - fv
    
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
(line1,) = ax.plot(x, f_(x))
(line2,) = ax.plot(x, -f_(x))


def update(a,b,c,d,e):
    line1.set_ydata(g_(x,a,b,c,d,e, True))
    line2.set_ydata(g_(x,a,b,c,d,e, False))
    fig.canvas.draw_idle()


interact(update, a=(-1, 1, 0.01), b=(-1, 1, 0.01), c=(-1, 1, 0.01), d=(-1, 1, 0.01), e=(-4, 4, 0.1));
```
